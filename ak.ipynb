```python
#converting intervals to integer values
Age = []
for i in data['age']:
  if i == '[80-90)':
    Age.append(85)
  elif i == '[60-70)':
    Age.append(65)
  elif i == '[50-60)':
    Age.append(55)
  elif i == '[70-80)':
    Age.append(75)
  elif i == '[40-50)':
    Age.append(45)
  elif i == '[20-30)':
    Age.append(25)
  elif i == '[30-40)':
    Age.append(35)
  elif i == '[90-100)':
    Age.append(95)
  elif i == '[0-10)':
    Age.append(5)
  else:
    Age.append(15)
```

Univariate analysis 
Univariate analysis of num_lab_procedures

```python
sns.FacetGrid(data,hue='readmitted', size=5 ).map(sns.distplot,'num_lab_procedures').add_legend()
```

```python
mn = data['num_lab_procedures'].mean()
md = data['num_lab_procedures'].median()
print('mean and median of num_lab_procedures is {:.2f} and {:.2f} respectively'.format(mn,md))
```

The distribution of number of lab procedures for both readmitted patients and non-readmitted patients is same but, as these features have considerable variation, they will prove to be useful for model to find patterns in overall complete training data.

```python
counts, bin_edges = np.histogram(data['num_lab_procedures'], bins=10, density=True)
pdf = counts/sum(counts)
cdf =np.cumsum(pdf)
plt.xlabel('num_lab_procedures')
plt.ylabel('likelihood')
plt.plot(bin_edges[1:], pdf)
plt.title('PMF of num_lab_procedures')
```

As seen from above pdf curve, majority of patients underwent 40 to 70 lab procedures with an average of around 43 and median of 44. The graph of readmitted and non-readmitted patients follows similar distribution for number_lab_procedures which is normally distributed.

```python
#Creating a Model 
x = data.drop('readmitted', axis=1)
y = data['readmitted']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=100)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf = DecisionTreeClassifier()
clf.fit(x_train, y_train)
clf.score(x_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
rfc.fit(x_train, y_train)
rfc.score(x_test, y_test)
```

```python
p_test =rfc.predict(x_test)
p_test
```

```python
from sklearn.metrics import f1_score
f1_score(y_test, p_test)
```
